```python
import re
#ÂÜô‰∏Ä‰∏™Ê∏ÖÊ¥óÂáΩÊï∞

def clean_text(text):
    # Áî®Á©∫Â≠óÁ¨¶‰∏≤ÊõøÊç¢ÊñáÊú¨‰∏≠ÊâÄÊúâ‰ª•"http"ÂºÄÂ§¥ÁöÑËøûÁª≠Â≠óÁ¨¶‰∏≤ÔºåÂπ∂Â∞ÜÊñáÊú¨ËΩ¨Êç¢‰∏∫Â∞èÂÜô
    text = re.sub("http\S+", "", text).lower()
    # ‰ΩøÁî®ÂàóË°®Êé®ÂØºÂºèÂ∞ÜÊñáÊú¨ÊãÜÂàÜ‰∏∫ÂçïËØçÂàóË°®ÔºåÂπ∂ËøáÊª§ÊéâÊâÄÊúâ‰ª•@ÂºÄÂ§¥ÁöÑÂçïËØç
    text = " ".join([w for w in text.split() if w[0]!='@'])
    # Áî®Á©∫Ê†ºÊõøÊç¢ÊñáÊú¨‰∏≠ÁöÑÊâÄÊúâÈùûÂ≠óÊØçÂíåÊï∞Â≠óÂ≠óÁ¨¶Ôºå‰ΩÜ‰∏çÊõøÊç¢Ê†áÁÇπÁ¨¶Âè∑ÔºåÂπ∂Â∞ÜÂâ©‰ΩôÁöÑÊñáÊú¨ËΩ¨Êç¢‰∏∫Â∞èÂÜô
    text = re.sub("[^a-zA-Z0-9\u4e00-\u9fa5,Ôºå.„ÄÇÔºü?ÔºÅ!;Ôºõ:]+", " ", text).lower()
    return text

test_text = 'hello @Apple, https://apple.com ÊùéJohn 12Ôºå34'
clean_text(text=test_text)
```

```python
#Ê∏ÖÊ¥ó‰∏Ä‰∏ã
import copy
data_clean1 = copy.deepcopy(data)#Ê∑±Êã∑Ë¥ù‰∏Ä‰∏™Êñ∞ÁöÑÂèòÈáèÁî®‰∫éÂ≠òÂÇ®Ê∏ÖÊ¥óÂêéÁöÑÊï∞ÊçÆ
data_clean1['Tweet_content'] = data_clean1['Tweet_content'].apply(clean_text)
data_clean1.dropna(subset=['Tweet_content'])
```

```python
import nltk

# Ê£ÄÊµã‰∏ÄÊÆµÊñáÊú¨‰∏≠ÁöÑËã±ÊñáÂçïËØç
text = data_clean1['Tweet_content'][51]
tokens = nltk.word_tokenize(text)

# ËøáÊª§ÊéâÊ†áÁÇπÁ¨¶Âè∑
words = [token for token in tokens if token.isalpha()]

# ËæìÂá∫Ê£ÄÊµãÂà∞ÁöÑÂçïËØç
print(words) 

len(words)
```

```python
#Âà†Èô§Â∞è‰∫éÁ≠â‰∫é5‰∏™ÂçïËØçÁöÑÊé®Êñá
data_clean2 = copy.deepcopy(data_clean1)
for index, row in data_clean2.iterrows():
    text = row['Tweet_content']
    tokens = nltk.word_tokenize(text)
    words = [token for token in tokens if token.isalpha()]
    if len(words) <=5:
        data_clean2.drop(index, inplace=True)
```
